# Data cleaning,checking for null value and filling it with the mean



```python
#checking to know if there is any null value
cardata.isnull().any()
```

```python
cardata.isnull().sum()
#cardata["EngineSize"].fillna(cardata["EngineSize"].mean())
# Calculate the mean of each column
#column_means = df.mean()

# Fill null values with the respective column means
#df_filled = df.fillna(column_means)
```

```python
# filling the numeric null values of each columns with the mean of the column
cardata.fillna(cardata.median(numeric_only=True), inplace=True)
```

```python
cardata.isnull().any()
```

```python
# filling the string column with the mode of the column
cardata.fillna(cardata.mode().iloc[0], inplace = True)
```
